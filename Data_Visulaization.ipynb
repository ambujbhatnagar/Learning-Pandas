# Creation of some basic graphs

### Setting up the environment 

```
import numpy as np 
import matplotlib.pyplot as plt 
```

```
# defining an array to be plotted 

xvals = np.arange(0,10,0.1)
```

Let us first plot a simple fucntion i.r. sin function output w.r.t to the xval array 

```
plt.plot(xvals, np.sin(xvals))
```

Now, the above graph can be modified to show two plots at the same time. Let us try by applying a cos function on the sin fuction and try to plot both at the same time 

```
plt.plot(xvals, np.sin(xvals))
plt.plot(xvals, np.cos(xvals))
```

if you wan to remove the gibberish code coming a the top of the plot. We can do that by putting a plt.show() function at the end of the code 

```
plt.plot(xvals, np.sin(xvals))
plt.plot(xvals, np.cos(xvals))
plt.show()
```

he next thing that one will like to do is label the chart for both the lines. This can be done using the xlabel and ylabel functions of plt libaray, Or the other way is by using the plt.legend() in our code.

The difference between the two is that. Xlabel and Ylabel functions will put a label on the x axis and y axis of the graph. Whilst the plt.legend() will put a label within the graph

```
plt.plot(xvals, np.sin(xvals), label = r'$y = \sin(x)$')
plt.plot(xvals, np.cos(xvals),  label = r'$y = \cos(x)$')
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

plt.show()
```

The next thing that we can think of is modfitying the colour of the lines. WHich can be done by putting a color = attritbute in the plt.plot()

```
plt.plot(xvals, np.sin(xvals), label = r'$y = \sin(x)$', color = 'red')
plt.plot(xvals, np.cos(xvals),  label = r'$y = \cos(x)$', color = 'green')
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

plt.show()
```

The above color codes can also be given in short form i.e. in RGB 

```
plt.plot(xvals, np.sin(xvals), label = r'$y = \sin(x)$', color = 'R')
plt.plot(xvals, np.cos(xvals),  label = r'$y = \cos(x)$', color = 'G')
plt.legend()
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

plt.show()
```

### Filtering graphs to focus on desired section 

```
plt.plot(xvals, np.sin(xvals), label = r'$y = \sin(x)$', color = 'R', linestyle = '--')
plt.plot(xvals, np.cos(xvals),  label = r'$y = \cos(x)$', color = 'G', linestyle = 'none', marker = 'o')
plt.legend()
plt.xlim(6,10)
plt.ylim(0,1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

plt.show()
```

There may be a lot of tiime when your plotted values are very dense and that means even a dot plot might seem like a solid line 

```
xvals = np.arange(0,10,0.01)

plt.plot(xvals, np.sin(xvals), marker = 'o' , linestyle = 'none')
plt.show()
```

The above problem can be solved using the markevery parameter of the plot function. This parameter will put a marker aver a user defined interval 

```
idx = 0
for i in [10,20,30]:
  plt.plot(xvals, np.sin(xvals) + idx , marker = 'o', linestyle = 'none' , markevery = i )
  idx += 1 
```

Another good option to use while plotting a line chart is markersize of the graph. 

```
idx = 0
for i , j in [[10,4],[20,8],[30,10]]:
  plt.plot(xvals, np.sin(xvals) + idx , marker = 'o', linestyle = 'none' , markevery = i , markersize = j)
  idx += 1 
```

### Ordering of the line 

One of the problem we face with multiple plot on same graph is that of overlapping which can be sorted out by zorder parameter of the plt.plot()

```
plt.plot(xvals, np.sin(xvals), lw = 5)
plt.plot(xvals + 1, np.sin(xvals), lw = 5)
plt.show()
```

What if we want to make the bluw line overlap the red line rather than otherway around 

```
plt.plot(xvals, np.sin(xvals), zorder = 1, lw = 5)
plt.plot(xvals + 1, np.sin(xvals), zorder = 0, lw = 5)
plt.show()
```

```
# Another way can be to use alpha parameter of matplotlib which makes the plot a bit opaue 
plt.plot(xvals, np.sin(xvals), zorder = 1, lw = 5, alpha = 0.5)
plt.plot(xvals + 1, np.sin(xvals), zorder = 0, lw = 5, alpha = 0.5)
plt.xlim(0,6)
plt.ylim(0,1)
plt.show()
```

### Jumbled up values 

Matplotlib has no way of identify how to connect the two values. It only follows the sequential values and connects them. In simple words, if the sequence of values are jumbleded up it may cause a web like plot to occure 

```
xvals = np.arange(0,10,0.1)
suffled = np.random.permutation(xvals) 
plt.plot(suffled , np.sin(suffled))
plt.show()
```

## Scatter plots 

had the above expmple been plotted on a scatter plot then the things will be different. And the reason for getting such a smooth plot is beccasue scatter plot only plots the markers 

```
xvals = np.arange(0,10,0.1)
suffled = np.random.permutation(xvals) 
plt.scatter(suffled , np.sin(suffled))
plt.show()
```

We can also change the colour of the markers using the parameter 'c' in the pt.scatter(). Another very interesting use of the c paramter will also be shown in the subsequent code section 

```
xvals = np.arange(0,10,0.1)
suffled = np.random.permutation(xvals) 
plt.scatter(suffled , np.sin(suffled), c = 'r')
plt.show()
```

```
xvals = np.arange(0,10,0.1)
suffled = np.random.permutation(xvals) 
plt.scatter(suffled , np.sin(suffled), c = np.cos(xvals))
plt.colorbar()
plt.show()
```

Now, the color map helps us introduce a third dimention to the data. The colorbar style can also be modified using the cmap parameter of the scatter plot function 

```
xvals = np.arange(0,10,0.1)
#suffled = np.random.permutation(xvals) 
plt.scatter(xvals , np.sin(xvals), c = np.cos(xvals), cmap = 'inferno', s = 100)
plt.colorbar()
plt.show()
```

Sometimes the border colour of the markers can cause the plot to look a bit out of place. This can be handled using the edgecolor parameter of scatter plot 

```
xvals = np.arange(0,10,0.1)
#suffled = np.random.permutation(xvals) 
plt.scatter(xvals , np.sin(xvals), c = np.cos(xvals), edgecolor = 'none', cmap = 'inferno', s = 100)
plt.colorbar()
plt.show()
```

```
xvals = np.arange(0,10,0.1)
#suffled = np.random.permutation(xvals) 
plt.scatter(xvals , np.sin(xvals), c = np.cos(xvals), cmap = 'inferno', s = np.power(xvals,3))
plt.colorbar()
plt.show()
```

## Bar plots 

Bar function doesn't take the simple positional arguments that plot or scatter does. Instead, you need to pass first as a positional argument, the x-axis, which in many cases tends to be an ordinal value rather than a numerical value. In other words, it's passing names for the bars rather than positions for the bars. And then, as the keyword argument height, the actual height of those bars.

```
nums = np.random.uniform(size = 10)
```

```
plt.bar(x = np.arange(0,10), height = nums)
```

The above graph is a bar graph with vertical bars and that is why we have given hieght as a parameter. However, if we want to create vertical bar graph we will need to input width of the colums 

```
plt.barh(np.arange(0,10), width = nums)
```

Now, again the colours of the  bar and the edgecolor can be changed in the plot using the color = ' ' parameter and edgecolor = ' ' parameter respectively. However, one tiny thing that I observe is that the bars are not centered around the values of the bar '

```
plt.barh(np.arange(0,10), width = nums, align = 'center')
```

Markers play a good role in bar charts when you want to distinguish between two different plots without changing the colors 

```
plt.barh(np.arange(0,10), width = nums, hatch = 'x', color = 'none', edgecolor = 'k')
plt.barh(np.arange(0,10) + 10, width = nums, hatch = 'o', color = 'none', edgecolor = 'g')
plt.show()
```

We can also modify the width of the bars using the width = parameter and also make the bars start from any other number other than the default 0.0 using the bottom = parameter. 
The bottom = parameter can also take a list type value as a parameter which helps in giving a different starting position to each bar. 

```
plt.bar(x = np.arange(0,10), height = nums, bottom = 0.1* nums)
```

## Representaion of 3 dimentional data 

3D data is where we have three information whcih we need to show on same graph. Foxample the density of customers and the distribution of customers within an area. Or average spend volume vs avg spend value with number of card per customer 

Let us take a very simple exaple of smooth gaussian distribution which makes things more relatable 

```
from scipy.ndimage.filters import gaussian_filter 
rand2d = gaussian_filter(np.random.normal(size = [512,512]), sigma = 10 )
print(rand2d.shape)
```

Do not be afraid with all the gibberish code written at top. A gaussian smooth distribution is a normal bell curve. It is just that we have created a normaal distribution in 2D with a signma 10. To imagine a 2D bell curve is difficult. This is why we are using the matplotlib to show you the third dimention 

```
plt.imshow(rand2d)
plt.colorbar()
plt.show()
```

```
plt.imshow(rand2d, cmap = 'magma')
plt.colorbar()
plt.show()
```
